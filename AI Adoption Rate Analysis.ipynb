```python
# Compare AI adoption rates (or revenue increase) between groups



import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from scipy.stats import ttest_ind

# Map regulation to numeric
reg_map = {'Lenient': 0, 'Moderate': 1, 'Strict': 2}
df['Regulation_Num'] = df['Regulation Status'].map(reg_map)

# Select columns for readiness components
readiness_features = ['AI-Generated Content Volume (TBs per year)', 
                      'Human-AI Collaboration Rate (%)', 
                      'Regulation_Num']

# Normalize features with Min-Max scaling
scaler = MinMaxScaler()
df_norm = df.copy()
df_norm[readiness_features] = scaler.fit_transform(df[readiness_features])

# Create composite readiness index (average of normalized components)
df_norm['Readiness_Index'] = df_norm[readiness_features].mean(axis=1)

# Median split for high and low readiness
median_readiness = df_norm['Readiness_Index'].median()
df_norm['Readiness_Group'] = np.where(df_norm['Readiness_Index'] >= median_readiness, 'High_Readiness', 'Low_Readiness')

# Group by regulation status (just use strict and lenient for clarity)
df_norm_filtered = df_norm[df_norm['Regulation Status'].isin(['Strict', 'Lenient'])]


# Now create groups combining regulation and readiness
df_norm_filtered.loc[:, 'Group'] = df_norm_filtered['Regulation Status'] + '_' + df_norm_filtered['Readiness_Group']

# Let's look at group sizes
print(df_norm_filtered['Group'].value_counts())

# Compare AI Adoption Rate across groups
grouped = df_norm_filtered.groupby('Group')['AI Adoption Rate (%)']

# Summary stats
print(grouped.agg(['mean', 'std', 'count']))

# T-test: High readiness + Strict vs Low readiness + Lenient
group1 = df_norm_filtered[(df_norm_filtered['Group']=='Strict_High_Readiness')]['AI Adoption Rate (%)']
group2 = df_norm_filtered[(df_norm_filtered['Group']=='Lenient_Low_Readiness')]['AI Adoption Rate (%)']

t_stat, p_val = ttest_ind(group1, group2, equal_var=False)

print(f"T-test Strict High Readiness vs Lenient Low Readiness: t={t_stat:.3f}, p={p_val:.3f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group data: mean and standard deviation of AI adoption rates per group
group_stats = df_norm_filtered.groupby('Group')['AI Adoption Rate (%)'].agg(['mean', 'std', 'count']).reset_index()

# Sort groups for readability
group_order = ['Strict_High_Readiness', 'Lenient_High_Readiness', 
               'Lenient_Low_Readiness', 'Strict_Low_Readiness']

# Create bar plot with error bars
plt.figure(figsize=(10, 6))
sns.barplot(
    data=group_stats,
    x='Group',
    y='mean',
    yerr=group_stats['std'],
    order=group_order,
    palette='Set2'
)

# Labels and formatting
plt.ylabel('Mean AI Adoption Rate (%)')
plt.xlabel('Group (Regulation + Readiness)')
plt.title('AI Adoption Rate by Regulation & Readiness Group')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

###   T-Test Result:
t = 0.996, p = 0.321

 Not statistically significant (since p > 0.05)



### Conclusion:
The hypothesis is partially supported — the average adoption is higher for strict + high readiness countries, but not enough to be statistically conclusive.

It’s still a valuable insight! It suggests readiness might matter more than regulation alone, but more data or refined segmentation could be needed to detect stronger effects.

### T_TEST Based on HIgh and Low AI Adoption rates:

```python
import pandas as pd
from  scipy.stats import ttest_ind



# divide into categories

high_threshold = df['AI Adoption Rate (%)'].quantile(0.75)
low_threshold = df['AI Adoption Rate (%)'].quantile(0.25)


# create groups

high_group = df[df['AI Adoption Rate (%)'] >= high_threshold]
low_group = df[df['AI Adoption Rate (%)'] <= low_threshold]

print(f"High group size: {len(high_group)}, Low group size: {len(low_group)}")
```

```python
# RUN T_TEST on key features



features_to_test = [
    'Revenue Increase Due to AI (%)',
    'Human-AI Collaboration Rate (%)',
    'Consumer Trust in AI (%)',
    'Job Loss Due to AI (%)',
    'AI-Generated Content Volume (TBs per year)'
]

# Store results
t_test_results = []

for feature in features_to_test:
    t_stat, p_value = ttest_ind(
        high_group[feature].dropna(),
        low_group[feature].dropna(),
        equal_var=False  # Welch's t-test for unequal variance
    )
    t_test_results.append({
        'Feature': feature,
        'High Group Mean': high_group[feature].mean(),
        'Low Group Mean': low_group[feature].mean(),
        'T-Statistic': t_stat,
        'P-Value': p_value
    })

# Display as DataFrame
results_df = pd.DataFrame(t_test_results)
print(results_df.sort_values('P-Value'))
```

### Strategic Takeaway:
There’s no strong statistical difference in key perception or outcome metrics (trust, revenue, collaboration, job loss) between high and low AI adopters.
This suggests:

Adoption levels might be driven more by infrastructure, strategy, or regulatory/cultural factors than these specific performance metrics.

Early adopters may not yet be seeing significantly different outcomes, supporting the idea that AI’s ROI and trust impact are still maturing.

```python
# Feature Importance via Regressor and Tree MOdel


from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

#select numeric_features

features = ['Revenue Increase Due to AI (%)', 'Human-AI Collaboration Rate (%)', 
            'Consumer Trust in AI (%)', 'Job Loss Due to AI (%)', 
            'AI-Generated Content Volume (TBs per year)']

target = ['AI Adoption Rate (%)']

X = df[features]
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)

model = RandomForestRegressor(n_estimators = 100, random_state=42)

model.fit(X_train, y_train.values.ravel())


#Feature Imporatnce

importances = pd.Series(model.feature_importances_, index=features).sort_values(ascending=False)
print(importances)
```

##### Interpretation & Insights
#####  Human-AI Collaboration is the top predictor

AI adoption thrives where humans and AI work together effectively, even under strict regulations.

Suggests a collaborative mindset or ecosystem is more influential than just trust or content output.

##### Revenue increase & content volume are nearly tied

Companies seeing financial benefits or using AI for content generation are adopting faster.
##### Job loss is not as negatively associated as assumed

Slightly counterintuitive: even with some job loss, companies may adopt AI aggressively — possibly for cost-cutting or productivity gains.

##### Consumer trust has the least impact

While still important, it's not the strongest driver in highly regulated regions — perhaps because these regions enforce minimum trust standards already.
